```python
def data2geojson(df, path_geojson):
    features = []
    insert_features = lambda X: features.append(
            geojson.Feature(geometry=geojson.Point((X[" X_opt"],
                                                    X[" Y_opt"],
                                                    X[" Z_opt"])),
                            properties=dict(imageName=X["imageName"],
                                           cameraName=X[" cameraName"],
                                           Omega=X[" Omega"],
                                           Phi=X[" Phi"],
                                            Kappa=X[" Kappa"]
                                           )))
    df.apply(insert_features, axis=1)
    with open(path_geojson, 'w', encoding='utf8') as fp:
        geojson.dump(geojson.FeatureCollection(features), fp, sort_keys=True, ensure_ascii=False)


data2geojson(df_txt, path_geojson)
```
